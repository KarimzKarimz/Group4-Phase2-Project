## Checking for Missing Values in Multiple Datasets

We will define a function to check for missing values in different DataFrames (`box_office_df`, `movie_basics_df`, `movie_rating_df`, and `film_df`) and display the columns with missing data.


```python
# Define a function to check for missing values in a DataFrame
def check_missing_values(df, name):
    print(f"\nüîç Missing values in {name}:")
    # Use .isnull() to identify missing data and .sum() to count them
    missing_values = df.isnull().sum()
    # Print only columns with missing values (where the sum is greater than 0)
    print(missing_values[missing_values > 0])  

# Check for missing values in each DataFrame
check_missing_values(box_office_df, "Box Office Data")
check_missing_values(film_df, "Merged Movie basics and Movie ratings Data")
check_missing_values(movie_budget_df, "movie budget Data")
```


- The function `check_missing_values(df, name)` checks for missing values in a given DataFrame:
  - `.isnull()` creates a DataFrame of boolean values (`True` for missing, `False` for not missing).
  - `.sum()` adds up the number of `True` values (i.e., missing entries) for each column.
  - We then filter and display only those columns with **missing values** (`missing_values > 0`).
- We run the function for each DataFrame to identify any missing data in the following:
  - **Box Office Data (`box_office_df`)**
  - **Movie Basics Data (`movie_basics_df`)**
  - **Movie Ratings Data (`movie_rating_df`)**
  - **Merged Movie Data (`film_df`)**

## Cleaning and Handling `foreign_gross` Column

We will:
- Remove any commas in the `foreign_gross` column.
- Convert the column values to `float`.
- Fill missing values (`NaN`) with the median value of the `foreign_gross` column.


```python
# Remove commas from 'foreign_gross' and convert it to a float
box_office_df["foreign_gross"] = box_office_df["foreign_gross"].replace(",", "", regex=True).astype(float)

# Fill missing values with the median of the 'foreign_gross' column
box_office_df["foreign_gross"].fillna(box_office_df["foreign_gross"].median(), inplace=True)

# Display the first few rows to check the results
print(box_office_df[['foreign_gross']].head())
```


- **`.replace(",", "", regex=True)`**: This removes any commas in the `foreign_gross` column to ensure that the values can be converted to numeric values.
- **`.astype(float)`**: After removing commas, we convert the column values to the `float` type for numerical operations.
- **`.fillna(box_office_df["foreign_gross"].median(), inplace=True)`**: Any missing values in the `foreign_gross` column are filled with the **median** value of the column. The median is often a good choice for filling missing values because it is less sensitive to outliers than the mean.
- We then print the first few rows of the `foreign_gross` column to check the changes.

## **HYPOTHESIS TESTING**

**Domestic vs Foreign Gross Revenue**

This project analyzes a movie dataset to test a hypothesis:
Is there a significant difference between domestic and foreign gross revenue of movies?

***Step 1: Importing Libraries***

scipy.stats.ttest_ind: to run an independent two-sample t-test


```python
#import necessary libraries
from scipy.stats import ttest_ind
```

***Step 2: Loading the Dataset***
Loads a CSV file named all_combined.csv that contains movie-related data.

```python
# load csv file
df = pd.read_csv("zippedData/all_combined.csv")
```

***Step 3: Cleaning the Data***

Strip "$" and "," from foreign_gross and domestic_gross_y columns.

Converts them to numeric data types so calculations can be done.

Invalid entries (e.g., strings or missing values) are set to NaN.

```python
#Clean the numeric columns
df['foreign_gross'] = pd.to_numeric(df['foreign_gross'].replace('[\$,]', '', regex=True), errors='coerce')
df['domestic_gross_y'] = pd.to_numeric(df['domestic_gross_y'].replace('[\$,]', '', regex=True), errors='coerce')

```

***Step 4: Defining Groups Based on Movie Ratings***

Splits the dataset into two groups:

High-rated movies: vote_average > 7

Low-rated movies: vote_average ‚â§ 7

```python
# Define groups based on rating
high_rating = df[df['vote_average'] > 7]
low_rating = df[df['vote_average'] <= 7]
```

***Step 5: Cleaning Revenue Data for Analysis***

Selects the worldwide_gross revenue column from both groups.

Removes missing (NaN) values to ensure the t-test runs correctly.

```python
# Drop missing values for revenue
high_revenue = high_rating['worldwide_gross'].dropna()
low_revenue = low_rating['worldwide_gross'].dropna()

```

***Step 6: Running the T-Test***

Performs an independent t-test to compare average worldwide revenue between high- and low-rated movies.


```python
# Run t-test
t_stat, p_value = ttest_ind(high_revenue, low_revenue, equal_var=False)
```

`equal_var=False` is used since the two groups may not have equal variances.

***Step 7: Displaying the Results***

Displays the mean revenue for each group.

Shows the t-statistic and p-value:

```python
# Print results
print("=== Hypothesis Test: High vs Low Rated Movies ===")
print(f"Mean Revenue (High Rating > 7): ${high_revenue.mean():,.2f}")
print(f"Mean Revenue (Low Rating ‚â§ 7): ${low_revenue.mean():,.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.5f}")
```

Since p_value < 0.05, we conclude there's a significant difference in revenue between the two rating groups.

***Step 8: Visualizing the Results***

Creates a bar chart to visually compare the average worldwide revenue of:

High-rated movies (green)

Low-rated movies (red)

Helps in easily identifying which group performs better financially.

```python
#Plot the comparison
plt.figure(figsize=(8, 5))
plt.bar(['High Rated', 'Low Rated'], [high_revenue.mean(), low_revenue.mean()], color=['green', 'red'])
plt.title('Average Revenue: High vs Low Rated Movies')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

#### ***Summary***

The project tests if movie ratings impact worldwide gross revenue.

A t-test was used to statistically compare the two groups (ratings >7 vs ‚â§7).

A significant difference (since p < 0.05) would suggest that high-rated movies earn more (or less) than low-rated ones.

A bar chart summarizes the findings visually.

## A.Impact of Budget on Box Office Revenue
`To evaluate whether investments on production budgets significantly affect box office performance`
- Null Hypothesis (H‚ÇÄ): Budget has no significant impact on box office revenue.
- Alternative Hypothesis (H‚ÇÅ): Higher budgets lead to significantly higher box office revenue

```python
#load the csv
budget_df=pd.read_csv('zippedData/cleaned_dataset_movie_budget.csv')
budget_df.head()
```

```python
budget_df.columns
```

```python
#rename worlwide_gross to revenue
budget_df=budget_df.rename(columns={'worldwide_gross':'revenue'})
print (budget_df.columns)
```

```python
#rename production_budget to budget
budget_df=budget_df.rename(columns={'production_budget':'budget'})
```

Proceed to check for missing values

```python
# Remove non-numeric characters (commas, dollar signs) and convert to numeric
for col in ['budget', 'revenue']:
    budget_df[col] = budget_df[col].astype(str).str.replace(r'[^\d.]', '', regex=True)
    budget_df[col] = pd.to_numeric(budget_df[col], errors='coerce')
```

```python
#drop rows with missing values
budget_df = budget_df.dropna(subset=['budget', 'revenue'])
```

Finalise the data cleaning .Assertaining the budget column has no Nan values

```python
# Force all to strings and lowercase
budget_df['budget'] = budget_df['budget'].astype(str).str.lower()

# Replace common bad values with NaN
budget_df['budget'] = budget_df['budget'].replace(['n/a', 'na', '-', 'none', 'unknown', '‚Äî'], pd.NA)

# Remove all non-numeric characters (except decimal point)
budget_df['budget'] = budget_df['budget'].str.replace(r'[^\d.]', '', regex=True)

# Convert to numeric
budget_df['budget'] = pd.to_numeric(budget_df['budget'], errors='coerce')


```

## Lets now Perform the T-Test


let‚Äôs `formulate the hypotheses`:

 We will need to obtain High budget and the Low budget threshhold by computation of the median production.This will split by the median 

```python
#COMPUTE THE MEDIAN PRODUCTION FOR THE BUDGET ANALYSIS
# Find the median budget
median_budget = budget_df['budget'].median()
print("Median budget:", median_budget)

# Split into high and low budget groups
low_budget_df = budget_df[budget_df['budget'] < median_budget]
high_budget_df = budget_df[budget_df['budget'] >= median_budget]

```

 Median Production Budget Analysis
The dataset was cleaned to extract valid numerical values for both budget and revenue. After removing any non-numeric characters and null entries, we computed the median production budget across all films.

Median Budget: $17,000,000

This median value serves as the threshold to divide the dataset into two groups for further statistical analysis:

Low Budget Films: Films with budgets ‚â§ $17M

High Budget Films: Films with budgets > $17M

This grouping is essential for hypothesis testing to determine whether higher production budgets are associated with significantly higher box office revenues.
We have assertained that the median value of 17,000,000 will be the threshhold of our statistical analysis.
Let us now calculate the T-statistic and P-value

```python
# Extract revenue
low_revenue = low_budget_df['revenue']
high_revenue = high_budget_df['revenue']

# Run Welch‚Äôs t-test
t_stat, p_value = ttest_ind(high_revenue, low_revenue, equal_var=False)

# Display results
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Interpretation
if p_value <= 0.05:
    print("Reject H‚ÇÄ: High-budget films earn significantly more revenue.")
else:
    print("Fail to reject H‚ÇÄ: No significant revenue difference based on budget.")
```
